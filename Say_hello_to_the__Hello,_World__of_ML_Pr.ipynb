Realice las operaciones dadas para una regresión lineal simple y visualizar los datos dados con **xs** y **ys**. Muestre el gráfico para analizar el comportamiento de los datos.

$Recordando$ $la$ $Regresión$ $Lineal$ $Simple:$

$ 
\overline{y} = β\overline{x} - α
$

$Donde$ $los$ $coeficientes:$

$ 
β = \frac{∑_{i=1}^nx_i∑_{i=1}^ny_i - n∑_{i=1}^nx_iy_i}{[∑_{i=1}^nx_í]^2 - n∑_{i=1}^nx_i^2}
$

$ 
α = \overline{y} - β\overline{x}
$

$Y$ $los$ $promedios:$

$ 
\overline{x} = ∑_{i=1}^n\frac{x_i}{n}
$

$ 
\overline{y} = ∑_{i=1}^n\frac{y_i}{n}
$

Entrene una red neuronal para hacer coincidir **xs** con **ys**. 

De modo que un predictor para un nuevo valor de X proporcione un valor flotante muy cercano a la respuesta deseada, es decir, **print(modelo.predict([X]))** debe dar un resultado satisfactorio.

```python
#Importaciones 

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
xs = np.array([-1.0,0.0,1.0,2.0,3.0,4.0], dtype=float)
ys = np.array([-4.0,1.0,6.0,11.0,16.0,21.0], dtype=float)

xs,ys
```

```python
from numpy.lib.function_base import average
# Funcion de la regresion lineal

def linear_regression(x=xs,y=ys,n=len(xs)):

  sumx = sum(x)
  sumy = sum(y)
  sumxy = sum((x*y))
  sumx2 = sum((x*x))

  promx = sum(x)/n
  promy = sum(y)/n

  beta = (sumx*sumy - n*sumxy)/(sumx**2 - n*sumx2)
  alpha = promy - beta*promx

  return beta,alpha
```

```python
linear_regression()
```

```python
b, a = linear_regression()
print(f'y={b}x+{a}')
```

```python
def linear_regression_visualization(beta = b, alpha = a):

  plt.plot(xs, ys, 'o', label = 'Datos')
  plt.plot(xs, beta*xs + alpha, label = 'Ajuste')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Say hello to the "Hello, World!" of ML')
  plt.grid()
  plt.legend()
  plt.show()

linear_regression_visualization()
```

```python
from tensorflow.python.ops.gen_array_ops import inplace_update
#Red neuronal con funcion de densidad y epocas o iteracciones

def solution_model():

  capa = tf.keras.layers.Dense(units=1,input_shape=[1])
  modelo = tf.keras.Sequential([capa])

  modelo.compile(optimizer='sgd', loss='mse')

  print('Comenzando entrenamiento...')
  historial = modelo.fit(xs,ys,epochs=500)
  print('Entrenamiento terminado')
  print('------------------------')

  resultado = modelo.predict([5.0])
  print('El resultado aproximado es: ' + str(resultado))

  return modelo,historial


  
```

```python
modelo,historial = solution_model()
modelo,historial
```

```python
def loss_function_visualization(historial = historial):
  plt.xlabel("Epocas")
  plt.ylabel("Magnitud de perdida")
  plt.plot(historial.history["loss"])

loss_function_visualization()
```
